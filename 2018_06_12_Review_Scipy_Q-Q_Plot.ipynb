```python
%matplotlib inline
# 균등분포 randn을 활용
np.random.seed(0)
x = np.random.randn(100) # 균등분포 uniform
plt.figure(figsize=(7, 7))
sp.stats.probplot(x, plot=plt)
plt.show()
```

```python
#정규 분포를 따르지 않는 데이터 샘플을 사용
np.random.seed(0)
x = np.random.rand(100)
plt.figure(figsize=(7, 7))
sp.stats.probplot(x, plot=plt)
plt.ylim(-0.5, 1.5)
plt.show()
```

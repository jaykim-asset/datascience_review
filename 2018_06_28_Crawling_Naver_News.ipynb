```python
from selenium import webdriver
import time

driver = webdriver.Chrome()
driver.get('http://news.naver.com/main/main.nhn?mode=LSD&mid=shm&sid1=105')

all_text = driver.find_elements_by_xpath('//*[@id="section_body"]/ul/li/dl/dt[2]/a')

ls = []

for span in all_text:
    ls.append(span.text)
    print(span.text)
for i in range(2, 11):
    driver.find_element_by_css_selector('#paging > a:nth-child({})'.format(i)).click()
    time.sleep(1)
    read_text = driver.find_elements_by_xpath('//*[@id="section_body"]/ul/li/dl/dt[2]/a')
    
    for span in read_text:
        ls.append(span.text)
        print(span.text)
```

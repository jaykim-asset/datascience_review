```python
# theta 가 0.6인 경우
%matplotlib inline
N = 10
theta = 0.6
rv = sp.stats.binom(N, theta)

xx = np.arange(N + 1) # 11개 0 ~ 10
plt.bar(xx, rv.pmf(xx), align='center')
plt.ylabel('P(x)')
plt.title('Binomial Distribution pmf')
plt.show()
```

```python
np.random.seed(0)
x = rv.rvs(100)
sns.countplot(x)
plt.title('Binomial Distribution Simulation')
plt.show()
```

```python
y = np.bincount(x, minlength=N+1)/float(len(x))
df = pd.DataFrame({'Theory': rv.pmf(xx), 'Simulation': y}).stack()
df = df.reset_index()
df.columns = ['values', 'type', 'ratio']
df.pivot('values', 'type', '')
```

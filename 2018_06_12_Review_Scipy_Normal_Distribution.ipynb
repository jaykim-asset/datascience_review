```python
# Normal Distribution 이론
%matplotlib inline
mu = 0
std = 1
rv = sp.stats.norm(mu, std)

xx = np.linspace(-5, 5, 100)
plt.plot(xx, rv.pdf(xx))
plt.ylabel('p(x)')
plt.title('Normal Distribution"s PDF')
plt.show()
```

```python
# Normal Distribution Simulation
np.random.seed(0)
x = rv.rvs(100)
sns.distplot(x, kde=True, fit=sp.stats.norm)
plt.show()
```

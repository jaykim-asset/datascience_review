#### Seaborn
    - http://seaborn.pydata.org/
    - Seaborn은 matplotlib을 기반으로 python시각화 라이브러리로 고급인터페이스를 제공하기 위한 패키지

```python
%matplotlib inline
%config InlineBackend.figure_formats = {'png', 'retina'}
```

```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
def sinplot():
    x = np.linspace(0, 10,101) # 0 ~ 10까지를 101등분함
    for i in range(6):
        plt.plot(x, np.sin(i +x)) # 60도씩 sin이 이동된 데이터를 그려줌
sinplot()
```

```python
# set seaborn
sns.set()
sinplot()
```

#### heatmap
- https://seaborn.pydata.org/generated/seaborn.heatmap.html

```python
tips = sns.load_dataset('tips') # tip Data 
tips.tail(5)
```

```python
tips_sum = tips.pivot_table(index='time', columns='sex', values='tip', aggfunc=np.sum)
tips_sum
```

```python
# annot: 수치 정보 출력, fmt: 수치데이터 포멧 ".2f" 소수 둘째 자리, as_cmap: 스무스하게 표현
# light_palette: http://seaborn.pydata.org/genearted/seaborn.light_palette.html
sns.heatmap(tips_sum, cmap=sns.light_palette('purple', as_cmap=True), annot=True, fmt=".2f")
```

```python
tips_mean = tips.pivot_table(index='time', columns='sex', values='tip', aggfunc=np.mean)
tips_mean
```

```python
# fmt: '.2g' 소수 둘째 반올림, 'd'정수
sns.heatmap(tips_mean, cmap=sns.light_palette('seagreen', as_cmap=True), annot=True, fmt='.2g')
```

```python
tips_mean = tips.pivot_table(index='time', columns='sex', values='tip', aggfunc=np.mean)
tips_mean
```

```python
sns.heatmap(tips_mean, cmap=sns.light_palette('seagreen', as_cmap=True), annot=True, fmt='0.2g')
```

```python
tips_sum = tips.pivot_table(index='time', columns='sex', values='tip', aggfunc=np.sum)
tips_sum
```

```python
sns.heatmap(tips_sum, cmap=sns.light_palette('seagreen', as_cmap=True), annot=True, fmt='0.2g')
```

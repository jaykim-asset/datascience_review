#### Multi line Drawing

```python
data1 = np.random.randint(5, size=5)
data2 = np.random.randint(5, size=5)
data3 = np.random.randint(5, size=5)
xticks = range(0,5)
print(data1, data2, data3)

plt.plot(xticks, data1, 'ro-', xticks, data2, 'bd:', xticks, data3, 'g^--')
plt.show()
```

```python
xticks = range(0, 5)
print(data1, data2, data3)

plt.plot(xticks, data1)
plt.plot(xticks, data2)
plt.plot(xticks, data3)
plt.show()
```

#### 여러개 그래프 출력

```python
data1 = np.random.randint(10, size=10)
data2 = np.random.randint(10, size=10)

plt.subplot(2,1,1) # 2 * 1 Matrix에서 1번째 위치
plt.plot(data1)

plt.subplot(2,1,2) # 2 * 1 Matrix에서 2번째 위치
plt.plot(data2)

plt.show()
```

```python
data1 = np.random.randint(20, size=20)
data2 = np.random.randint(10, size=10)
data3 = np.random.randint(15, size=15)
data4 = np.random.randint(20, size=20)

plt.figure(figsize=(20,10)) # Size 20 , 4 -> 가로확대 그림
plt.subplot(2,2,1) # 2x2 행렬의 1번째 그림
plt.xlim(0, 20) # x축값 0 ~ 20 으로 제한
plt.ylim(0, 20) # y축값 0 ~ 20 으로 제한
plt.xticks(range(20+1)) # x ticks 0 ~ 19까지
plt.yticks(range(20+1)) # y ticks 0 ~ 19까지
plt.xlabel('price1') # x축 이름: price1
plt.ylabel('quantity1') # y축 이름: quantity 1
plt.title('KRX1') # 제목: KRX1
plt.legend(loc=0) # legend best fit 설정
plt.plot(data1, 'r--') # 빨간 -- 선으로 그림

plt.figure(figsize=(20,10))
plt.subplot(2,2,2)
plt.xlim(0, 10)
plt.ylim(0, 10)
plt.xticks(range(10+1))
plt.yticks(range(10+1)) # Range로 설정
plt.xlabel('price2')
plt.ylabel('quantity2')
plt.title('KRX2')
plt.legend(loc=0)
plt.plot(data2, 'b:')

plt.figure(figsize=(20,10))
plt.subplot(2,2,3)
plt.xlim(0, 15)
plt.ylim(0, 15)
plt.xticks(range(15+1))
plt.yticks(range(15+1))
plt.xlabel('price3')
plt.ylabel('quantity3')
plt.title('KRX3')
plt.legend(loc=0)
plt.plot(data3, 'g--')

plt.figure(figsize=(20,10))
plt.subplot(2,2,4)
plt.xlim(0, 20)
plt.ylim(0, 20)
plt.xticks(range(20+1))
plt.yticks(range(20+1))
plt.xlabel('price4')
plt.ylabel('quantity4')
plt.title('KRX4')
plt.legend(loc=0)
plt.plot(data4, 'p-')

plt.show()
```

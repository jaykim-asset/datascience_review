```python
def make_model(i, station):
    d_variables = "np.log1p(units) ~ " + "".join(stations_var_c[i-1]) + "+0"
    model1 = sm.OLS.from_formula(d_variables, data = station)
    result1 = model1.fit()

    # 아웃라이어 제거
    # Cook's distance > 2 인 값 제거
    influence = result1.get_influence()
    cooks_d2, pvals = influence.cooks_distance
    fox_cr = 4 / (len(station) - 2)
    idx_outlier = np.where(cooks_d2 > fox_cr)[0]

    idx = list(set(range(len(station))).difference(idx_outlier))
    station_1 = station.iloc[idx, :].reset_index(drop=True)

    model2 = sm.OLS.from_formula(d_variables, station_1)
    result2 = model2.fit()
    b = data_set[i-1]
    predict_value = result2.predict(b)
    b["y"] = predict_value
    
    return data_set[i-1]
```

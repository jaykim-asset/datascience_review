```python
new_0 = pd.to_numeric(data_weather['tmax'], errors = 'coerce')
new_1 = pd.to_numeric(data_weather['tmin'], errors = 'coerce')
new_2 = pd.to_numeric(data_weather['tavg'], errors = 'coerce')
new_3 = pd.to_numeric(data_weather['depart'], errors = 'coerce')
new_4 = pd.to_numeric(data_weather['dewpoint'], errors = 'coerce')
new_5 = pd.to_numeric(data_weather['wetbulb'], errors = 'coerce')
new_6 = pd.to_numeric(data_weather['heat'], errors = 'coerce')
new_7 = pd.to_numeric(data_weather['cool'], errors = 'coerce')
new_8 = pd.to_numeric(data_weather['snowfall'], errors = 'coerce')
new_9 = pd.to_numeric(data_weather['preciptotal'], errors = 'coerce')
new_10 = pd.to_numeric(data_weather['stnpressure'], errors = 'coerce')
new_11 = pd.to_numeric(data_weather['sealevel'], errors = 'coerce')
new_12 = pd.to_numeric(data_weather['resultspeed'], errors = 'coerce')
new_13 = pd.to_numeric(data_weather['resultdir'], errors = 'coerce')
new_14 = pd.to_numeric(data_weather['avgspeed'], errors = 'coerce')

data_weather['tmax'] = new_0.combine_first(data_weather['tmax'])
data_weather['tmin'] = new_1.combine_first(data_weather['tmin'])
data_weather['tavg'] = new_2.combine_first(data_weather['tavg'])
data_weather['depart'] = new_3.combine_first(data_weather['depart'])
data_weather['dewpoint'] = new_4.combine_first(data_weather['dewpoint'])
data_weather['wetbulb'] = new_5.combine_first(data_weather['wetbulb'])
data_weather['heat'] = new_6.combine_first(data_weather['heat'])
data_weather['cool'] = new_7.combine_first(data_weather['cool'])
data_weather['snowfall'] = new_8.combine_first(data_weather['snowfall'])
data_weather['preciptotal'] = new_9.combine_first(data_weather['preciptotal'])
data_weather['stnpressure'] = new_10.combine_first(data_weather['stnpressure'])
data_weather['sealevel'] = new_11.combine_first(data_weather['sealevel'])
data_weather['resultspeed'] = new_12.combine_first(data_weather['resultspeed'])
data_weather['resultdir'] = new_13.combine_first(data_weather['resultdir'])
data_weather['avgspeed'] = new_14.combine_first(data_weather['avgspeed'])
```

```python
# tmax mean
max_avg = data_weather['tmax'][data_weather['tmax'] != 'M'].mean().round(2)
# tmin mean
min_avg = data_weather['tmin'][data_weather['tmin'] != 'M'].mean().round(2)
# tavg mean
avg_avg = data_weather['tavg'][data_weather['tavg'] != 'M'].mean().round(2)
# depart mean
depart_avg = data_weather['depart'][data_weather['depart'] != 'M'].mean().round(2)
# dewpoint mean
dewpoint_avg = data_weather['dewpoint'][data_weather['dewpoint'] != 'M'].mean().round(2)
# wetbulb mean
wetbulb_avg = data_weather['wetbulb'][data_weather['wetbulb'] != 'M'].mean().round(2)
# heat mean
heat_avg = data_weather['heat'][data_weather['heat'] != 'M'].mean().round(2)
# cool mean
cool_avg = data_weather['cool'][data_weather['cool'] != 'M'].mean().round(2)
# snowfall mean
snowfall_avg = data_weather['snowfall'][(data_weather['snowfall'] != 'M') & (data_weather['snowfall'] != '  T')].mean().round(2)
# preciptotal mean
preciptotal_avg = data_weather['preciptotal'][(data_weather['preciptotal'] != 'M') & (data_weather['preciptotal'] != '  T')].mean().round(2)
# stnpressure mean
stnpressure_avg = data_weather['stnpressure'][data_weather['stnpressure'] != 'M'].mean().round(2)
# sealevel mean
sealevel_avg = data_weather['sealevel'][data_weather['sealevel'] != 'M'].mean().round(2)
# resultspeed mean
resultspeed_avg = data_weather['resultspeed'][data_weather['resultspeed'] != 'M'].mean().round(2)
# resultdi mean
resultdir_avg = data_weather['resultdir'][data_weather['resultdir'] != 'M'].mean().round(2)
# avgspeed mean
avgspeed_avg = data_weather['avgspeed'][data_weather['avgspeed'] != 'M'].mean().round(2)

```

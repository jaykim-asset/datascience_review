```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

#### Bar plot

```python
y = np.random.randint(10, size=5)
x = range(1, 6)
plt.bar(x, y)
plt.show()
```

```python
y  = np.random.randint(5, 10, size=5)
x = range(1, 6)
error = np.random.randint(1, 5, size=5)
print(y, error)
plt.bar(x, y, yerr=error, align='center', alpha = 0.5)
plt.show()
```

```python
y = np.random.randint(5, 10, size=5)
x = range(1,6)
error = np.random.randint(1, 5, size=5)
print(y, error)
plt.barh(x, y, xerr=error, align='center', alpha=0.8, color='r')
plt.show()
```

```python
y1 = np.random.randint(5, 10, size=5)
y2 = np.random.randint(1, 5, size=5)
print(y1, y2)
plt.plot()
x = range(1, 6)
plt.bar(x, y1, color='r')
plt.bar(x, y2, color='y')
plt.show()
```

#### pie plot

```python
data = np.random.randint(1, 6, size= 5)
labels = ['KM3', 'KM10', 'K200F', 'MiniK200F','Option']
colors = ['yellowgreen', 'gold', 'lightskyblue', 'lightcoral', 'red']
print(data)
plt.pie(data, labels=labels, colors=colors, autopct='%1.2f%%')
plt.show()
```

#### scatter plot

```python
# uniform(균등분포)
x1 = np.random.rand(500)
y1 = np.random.rand(500)
plt.subplot(331)
plt.scatter(x1, y1)

# gaussian (정규분포)
x2 = np.random.randn(500)
y2 = np.random.randn(500)
plt.subplot(332)
plt.scatter(x2, y2)

# randint (uniform)
x3 = np.random.randint(1, 100, size=500)
y3 = np.random.randint(1, 100, size=500)
plt.subplot(333)
plt.scatter(x3, y3)

plt.show()
```

```python
data1 = np.random.randint(1, 10, size=20)
data2 = np.random.randint(1, 10, size=20)
colors= np.random.rand(20)
area = np.pi * np.arange(10, 50, 2) ** 2
plt.scatter(data1, data2, s=area, c=colors, alpha=0.3)
plt.show()
```

##### histogram

```python
# 50개의 구간을 잘라 해당되는 데이터의 수를 카운터한 후, y축의 값을 normalized함.
x1 = np.random.rand(1000) # 균등분포를 보임, rand를 써서 그럼
plt.figure(figsize=(8,2))
plt.subplot(2,1,1)
plt.hist(x1, bins=50)
plt.show()

x2 = np.random.randn(1000) # 정규분포를 보임 randn을 써서 그럼
plt.figure(figsize=(20,2))
plt.subplot(2,2,2)
plt.hist(x2, bins=50)
plt.show()
```
